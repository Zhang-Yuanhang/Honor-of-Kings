```python
def combination_win_rate(hero1,hero2):
    id1 = hero_lst[hero_lst['heroName'] == hero1].index[0]
    id2 = hero_lst[hero_lst['heroName'] == hero2].index[0]
    win_games = Win_list_final[(Win_list_final[['win_1','win_2','win_3','win_4','win_5']] == str(id1)).any(axis=1) & (Win_list_final[['win_1','win_2','win_3','win_4','win_5']] == str(id2)).any(axis=1)].shape[0]
    lose_games = Win_list_final[(Win_list_final[['lose_1','lose_2','lose_3','lose_4','lose_5']] == str(id1)).any(axis=1) & (Win_list_final[['lose_1','lose_2','lose_3','lose_4','lose_5']] == str(id2)).any(axis=1)].shape[0]
    return win_games / (win_games+lose_games), win_games+lose_games
```

```python
sho_lst = hero_lst[hero_lst['position'] == '边']['heroName'].values
ass_lst = hero_lst[hero_lst['position'] == '辅']['heroName'].values
```

```python
sho_ass_win_matrix = pd.DataFrame()
sho_ass_games_matrix = pd.DataFrame()

for sho in sho_lst:
    for ass in ass_lst:
        rate, games = combination_win_rate(sho,ass)
        sho_ass_win_matrix.loc[sho,ass] = rate
        sho_ass_games_matrix.loc[sho,ass] = games
```

```python
sho_ass_matrix = pd.concat([sho_ass_win_matrix.unstack().to_frame(),
sho_ass_games_matrix.unstack().to_frame()],axis=1)
sho_ass_matrix.columns = ['胜率','胜场']
```

```python
Win_lst_all = pd.DataFrame()
for win in ['win_1','win_2','win_3','win_4','win_5']:
    for lose in ['lose_1','lose_2','lose_3','lose_4','lose_5']:
        Win_lst = Win_list_final[[win,lose]]
        Win_lst.columns = ['win', 'lose']
        Win_lst_all = pd.concat([Win_lst_all, Win_lst])
```

```python
Win_lst_all_ally = pd.DataFrame()
for win1 in range(5-1):
    for win2 in range(win1+1,5):
        Win_lst = Win_list_final[[['win_1','win_2','win_3','win_4','win_5'][win1],['win_1','win_2','win_3','win_4','win_5'][win2]]]
        Win_lst.columns = ['ally1', 'ally2']
        Win_lst_all_ally = pd.concat([Win_lst_all_ally, Win_lst])

Lose_lst_all_ally = pd.DataFrame()
for lose1 in range(5-1):
    for lose2 in range(lose1+1,5):
        Lose_lst = Win_list_final[[['lose_1','lose_2','lose_3','lose_4','lose_5'][lose1],['lose_1','lose_2','lose_3','lose_4','lose_5'][lose2]]]
        Lose_lst.columns = ['ally1', 'ally2']
        Lose_lst_all_ally = pd.concat([Lose_lst_all_ally, Lose_lst])
```

```python
Win_lst_all['games'] = 1
```

```python
Win_lst_all_ally['games'] = 1
Lose_lst_all_ally['games'] = 1
```

```python
Win_lst_ally = Win_lst_all_ally.groupby(['ally1', 'ally2']).count().reset_index().iloc[:-1]
Lose_lst_ally = Lose_lst_all_ally.groupby(['ally1', 'ally2']).count().reset_index().iloc[:-1]
```

```python
Win_ally_matrix = Win_lst_ally.pivot(index='ally1',columns='ally2',values='games').fillna(0)
Lose_ally_matrix = Lose_lst_ally.pivot(index='ally1',columns='ally2',values='games').fillna(0)
```

```python
Win_ally_matrix =  Win_ally_matrix + Win_ally_matrix.T
Lose_ally_matrix = Lose_ally_matrix + Lose_ally_matrix.T
```

```python
Win_lst_final = Win_lst_all.groupby(['win','lose']).count().reset_index().iloc[:-1]
```

```python
Win_games_matrix = Win_lst_final.pivot(index='win',columns='lose',values='games').fillna(0)
```

```python
hero_name_lst = hero_lst.reset_index().sort_values(by='heroId')['heroName'].values
```

```python
Win_rate_matrix = Win_games_matrix.div(Win_games_matrix + Win_games_matrix.T)
Win_rate_matrix = Win_rate_matrix.fillna(0).round(4)
```

```python
Win_rate_ally_matrix = Win_ally_matrix.div(Win_ally_matrix + Lose_ally_matrix)
Win_rate_ally_matrix = Win_rate_ally_matrix.fillna(0).round(4)
```

```python
Win_games_matrix.index = hero_name_lst 
Win_games_matrix.columns = hero_name_lst 

Win_rate_matrix.index = hero_name_lst 
Win_rate_matrix.columns = hero_name_lst 

Win_ally_matrix.index = hero_name_lst 
Win_ally_matrix.columns = hero_name_lst 

Lose_ally_matrix.index = hero_name_lst 
Lose_ally_matrix.columns = hero_name_lst

Win_rate_ally_matrix.index = hero_name_lst 
Win_rate_ally_matrix.columns = hero_name_lst 
```

```python
## 胜率根据胜场进行调整
# adj_rate_oppo = np.sqrt(np.sqrt((Win_games_matrix + Win_games_matrix.T)/(Win_games_matrix + Win_games_matrix.T).max().max()))
# adj_rate_ally = np.sqrt(np.sqrt((Win_ally_matrix + Lose_ally_matrix)/(Win_ally_matrix + Lose_ally_matrix).max().max()))

adj_rate_oppo = (Win_games_matrix + Win_games_matrix.T)/(Win_games_matrix + Win_games_matrix.T).max().max()
adj_rate_ally = (Win_ally_matrix + Lose_ally_matrix)/(Win_ally_matrix + Lose_ally_matrix).max().max()
```

```python
Win_rate_matrix_adj = 0.5 + (Win_rate_matrix - 0.5) * adj_rate_oppo
Win_rate_ally_matrix_adj = 0.5 + (Win_rate_ally_matrix - 0.5) * adj_rate_ally
```

```python
## 根据胜场惩罚得到的最新推荐系统
def hero_commendation(allys=['公孙离','少司缘'], enemys=['马可波罗','瑶'], pos=None):
    if pos:
        try:
            pos_lst = hero_lst[hero_lst['position'] == pos]['heroName'].values
            Win_matrix = Win_rate_matrix_adj.loc[pos_lst]
            Win_matrix_ally = Win_rate_ally_matrix_adj.loc[pos_lst]
        except:
            Win_matrix = Win_rate_matrix_adj
            Win_matrix_ally = Win_rate_ally_matrix_adj
    else:
        Win_matrix = Win_rate_matrix_adj
        Win_matrix_ally = Win_rate_ally_matrix_adj
    res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).sort_values(ascending=False).head(20)
    return res
```
