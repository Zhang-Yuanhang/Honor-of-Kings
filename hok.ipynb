```python
def combination_win_rate(hero1,hero2):
    id1 = hero_lst[hero_lst['heroName'] == hero1].index[0]
    id2 = hero_lst[hero_lst['heroName'] == hero2].index[0]
    win_games = Win_list_final[(Win_list_final[['win_1','win_2','win_3','win_4','win_5']] == str(id1)).any(axis=1) & (Win_list_final[['win_1','win_2','win_3','win_4','win_5']] == str(id2)).any(axis=1)].shape[0]
    lose_games = Win_list_final[(Win_list_final[['lose_1','lose_2','lose_3','lose_4','lose_5']] == str(id1)).any(axis=1) & (Win_list_final[['lose_1','lose_2','lose_3','lose_4','lose_5']] == str(id2)).any(axis=1)].shape[0]
    return win_games / (win_games+lose_games), win_games+lose_games
```

```python
## 胜率根据胜场进行调整
# adj_rate_oppo = np.sqrt(np.sqrt((Win_games_matrix + Win_games_matrix.T)/(Win_games_matrix + Win_games_matrix.T).max().max()))
# adj_rate_ally = np.sqrt(np.sqrt((Win_ally_matrix + Lose_ally_matrix)/(Win_ally_matrix + Lose_ally_matrix).max().max()))

adj_rate_oppo = (Win_games_matrix + Win_games_matrix.T)/(Win_games_matrix + Win_games_matrix.T).max().max()
adj_rate_ally = (Win_ally_matrix + Lose_ally_matrix)/(Win_ally_matrix + Lose_ally_matrix).max().max()
```

```python
def hero_commendation(allys=['公孙离','少司缘'], enemys=['马可波罗','瑶'], pos='野', adj=None):
    if pos:
        try:
            # pos_lst = hero_lst[hero_lst['position'] == pos]['heroName'].values
            pos_lst = ['镜','韩信','裴擒虎','云中君','诸葛亮','司马懿','宫本武藏','暃','马超','典韦','猪八戒','雅典娜','坦元','夏侯惇','百里玄策']
            if adj:
                Win_matrix = Win_rate_matrix_adj.loc[pos_lst]
                Win_matrix_ally = Win_rate_ally_matrix_adj.loc[pos_lst]
            else:
                Win_matrix = Win_rate_matrix.loc[pos_lst]
                Win_matrix_ally = Win_rate_ally_matrix.loc[pos_lst]
            res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[pos_lst].sort_values(ascending=False).head(3)
            res = res.to_frame('win_rate').reset_index()
        except:
            pass
    else:
        if adj:
            Win_matrix = Win_rate_matrix_adj
            Win_matrix_ally = Win_rate_ally_matrix_adj
        else:
            Win_matrix = Win_rate_matrix
            Win_matrix_ally = Win_rate_ally_matrix

        edg_res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[edg_lst].sort_values(ascending=False).head(3)
        jun_res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[jun_lst].sort_values(ascending=False).head(3)
        mid_res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[mid_lst].sort_values(ascending=False).head(3)
        sho_res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[sho_lst].sort_values(ascending=False).head(3)
        ass_res = pd.concat([Win_matrix_ally[allys],Win_matrix[enemys]],axis=1).mean(axis=1).loc[ass_lst].sort_values(ascending=False).head(3)

        edg_res = edg_res.to_frame('win_rate').reset_index()
        edg_res.insert(0,'position','边')
        jun_res = jun_res.to_frame('win_rate').reset_index()
        jun_res.insert(0,'position','野')
        mid_res = mid_res.to_frame('win_rate').reset_index()
        mid_res.insert(0,'position','中')
        sho_res = sho_res.to_frame('win_rate').reset_index()
        sho_res.insert(0,'position','射')
        ass_res = ass_res.to_frame('win_rate').reset_index()
        ass_res.insert(0,'position','辅')

        res = pd.concat([edg_res,jun_res,mid_res,sho_res,ass_res],axis=0).set_index(['position','index'])
    return res
```

```python
def hero_commendation_score(allys=['关羽','王昭君','张良'], enemys=['马可波罗','瑶'], pos='野', adj=None):
    
    enemy_weight = {
        '边': 0.2,
        '野': 0.25,
        '中': 0.2,
        '射': 0.25,
        '辅': 0.1
    }
    # pos_lst = hero_lst[hero_lst['position'] == pos]['heroName'].values
    # pos_lst = ['镜','韩信','裴擒虎','云中君','诸葛亮','司马懿','宫本武藏','暃','马超','典韦','猪八戒','雅典娜','坦元','夏侯惇','百里玄策']
    pos_lst = ['镜','杨戬','裴擒虎','诸葛亮','司马懿','宫本武藏','暃','韩信','铠','澜','曜','大司命']
    if adj:
        Win_matrix = Win_rate_matrix_adj.loc[pos_lst]
        Win_matrix_ally = Win_rate_ally_matrix_adj.loc[pos_lst]
    else:
        Win_matrix = Win_rate_matrix.loc[pos_lst]
        Win_matrix_ally = Win_rate_ally_matrix.loc[pos_lst]
    
    # RankIC 调整：
    Win_matrix_ally_rank = Win_matrix_ally.rank(ascending=True,axis=0)
    Win_matrix_rank = Win_matrix.rank(ascending=True,axis=0)

    res_enemy = pd.DataFrame()
    for enemy in enemys:
        print(enemy)
        enemy_pos = hero_lst.query(f"heroName == '{enemy}'")['position'].values[0]
        weight = enemy_weight[enemy_pos]
        res_enemy = pd.concat([res_enemy, Win_matrix_rank[enemy]*weight],axis=1)

    res_ally = Win_matrix_ally_rank[allys]*0.1
    res_enemy = res_enemy

    res = pd.concat([res_ally,res_enemy],axis=1)
    res['Final_Score'] = res.sum(1)
    
    res = res.sort_values(by = 'Final_Score',ascending=False).head(5)
    return res
```
